```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = "/content/drive/MyDrive/Data Mining Project/data/unclean data/FD_INDREGZA_2016.csv"

# Read the CSV file into a Pandas DataFrame
datazone = pd.read_csv(csv_file_path, delimiter=";")

datazone
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the list of documents
df = datazone

# Columns to keep for plotting frequencies
columns_to_keep = ['STAT_CONJ']

# Plot frequencies for each column
for col in columns_to_keep:
    plt.figure(figsize=(8, 6))
    df[col].value_counts().plot(kind='bar', title=f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the list of documents
df = datazone

# Columns to keep for plotting frequencies
columns_to_keep = ['STAT_CONJ', 'APAF', 'LPRM', 'INFAM', 'COUPLE', 'NPERR', 'NAT49', 'NATN12', 'IMMI', 'INAT', 'DIPL_15', 'STAT']

# Values to exclude from conversion to int
exclude_values = ["Z", "ZZ", "A", "B", "C", "D", "AFR", "AOC", "CAR", "EUR", "HUE", "AME"]

# Plot frequencies for each column
for col in columns_to_keep:
    plt.figure(figsize=(8, 6))

    # Convert all values to int except the excluded ones
    df[col] = df[col].apply(lambda x: int(x) if x not in exclude_values else x)

    df[col].value_counts().plot(kind='bar', title=f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
