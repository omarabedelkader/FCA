```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the list of documents
df = datazone

# Columns to keep for plotting frequencies
columns_to_keep = ['STAT_CONJ', 'APAF', 'LPRM', 'INFAM', 'COUPLE', 'NPERR', 'NAT49', 'NATN12', 'IMMI', 'INAT', 'DIPL_15', 'STAT']

# Values to exclude from conversion to int
exclude_values = ["Z", "ZZ", "A", "B", "C", "D", "AFR", "AOC", "CAR", "EUR", "HUE", "AME"]

# Minimum frequency threshold
min_frequency = 1000

# Change the type of unique values to int except excluded values
for col in columns_to_keep:
    unique_values = df[col].unique()
    for value in unique_values:
        if value not in exclude_values:
            df[col] = df[col].replace(value, int(value))

# Calculate the number of rows and columns for subplots
num_columns = 3
num_rows = len(columns_to_keep) // num_columns + (len(columns_to_keep) % num_columns > 0)

# Create subplots
fig, axs = plt.subplots(num_rows, num_columns, figsize=(30, 20))

# Flatten the 2D axs array for easy iteration
axs = axs.flatten()

# Plot frequencies for each column
for i, col in enumerate(columns_to_keep):
    value_counts = df[col].value_counts()

    # Filter values based on frequency
    filtered_value_counts = value_counts[value_counts >= min_frequency]

    filtered_value_counts.plot(kind='bar', ax=axs[i])
    axs[i].set_title(f'Frequency of {col}')
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Frequency')

# Remove any empty subplots if there are fewer columns than rows * columns
for i in range(len(columns_to_keep), num_rows * num_columns):
    fig.delaxes(axs[i])

plt.tight_layout()
plt.show()
```
